# **Lecture 2 and 3: variables a programmers perspective and dynamic typing**

```python
# In this session we are gonna learn about dynamic typing
""" What is dynamic typing
It is a property of variable. It states that a variable containing value can be changed by the property
To understand it more clearly, let's see some examples """
a = 15
print(type(a))

a = 'Homo sepians'
print(type(a))
# as we can see that first variable contains the integer value, then it stored the string value, and this property of variables are called dynamic typing
# Let's see some more examples of dynamic typing.
b = 18
print(type(b))
print(b)
b = 18/2
print(type(b))
print(b)
```

# **Lecture 5: Escape characters and types of quotes**

*   Escape character <\\> the following symbol is called escape character it can be used for many ways.

1.   <\t> it's called escape tab which is used to give extra spaces in sentences.
2.   <\n> this is new line tab which breaks the line in sentence and print it on new line.



*  Quotes (") which is used to stores variable in string form it is of 3 types
(''), (" ") these are single quotes and double quotes. now we are gonna learn about triple quotes which is given by (""" """).













---



```python
# Escape character let's see an example to understand it...
# print('It's a beautiful day')              # On printing it will show an error because there extra quotes which which is not readable by python .
# but by using <\> this escape character it will form an special character .
print('It\'s a beautiful day')

# Now, if we want the space between lines we use <\t> tab escape to make spaces ..
print('This is my name, and this world revolve around it.')
print('This is my name,\tand this world revolve around it.')
# we can use tab more often. eg,
print('This is my name,\t\tand this world revolve around it.')

# new escape line <\n>
print('Hello everyone. How is day on earth.')
print('Hello everyone. \nHow is day on earth.')

# Quotes
x = 'Hello People of world.'
y = "How are you all."
z = """ First line
second line
third line
Fourth line"""
f = '''mango
apple
pineapple '''          # This is triple quotes which is used to store string in multiple lines and it can be done in single or double quotes.
print(x)
print(y)
print(z)
print(f)

# let's see something interesting ..
print("It's a beautiful day")
print('we are from "IIT" madras')         # It prints by changing the quotes .
```

```python
from ast import Yield
x = "pyTHoN sTring mEthOdS"
print(x.lower())        # It make all the string in lower case letter
print(x.upper())          # It make all the string in upper case letter
print(x.capitalize())     # It make first letter of first word capital
print(x.title())          # converts the first character of each word to upper case.
print(x.swapcase())     # It swaps cases, lower case become upper case and vice versa.

a = input("Enter a string:\n ")
print(a.lower())
print(a.upper())
print(a.capitalize())
print(a.title())
print(a.swapcase())

# To check string methods we use "is" operator, which are of different types to check the function. It always gives result in boolean value.
# 1). "islower()" function to check the string is in lower case or not.
r = 'string'
s = 'String'
print(r.islower())
print(s.islower())

# 2). "isupper()" function is used to check string is upper case or not.
q = 'FOOD'
r_ = 'Apple'
print(q.isupper())
print(r_.isupper())

# 3). "istitle()" function is used to check string value is title or not.
x1 = "Apples can be green."
x2 = "Apples Can Be Green."
print(x1.istitle())
print(x2.istitle())

# 4). "isdigit()" is function which return true when all characters in string is digit.
x3 = '46'
x4 = '45av'
print(x3.isdigit())
print(x4.isdigit())

# 5). "isalpha()" is function which return true when all characters in string is alphabet.
x5 = "pineapple44"
x6 = 'jackfruit'
print(x5.isalpha())
print(x6.isalpha())

# 6). "isalnum()" is function which return true if all characters in string is alpha numeric.
x7 = 'abracadabra@'
x8 = 'abra7cadabra'
print(x7.isalnum())
print(x8.isalnum())

# Strip operator is methods in string function which strips extra spaces, characters, dashes, etc.
# there are of total 3 strips methods.
# 1). "strip" this function is used to strip extra spaces on both sides.
y = '-----Dakota-----'
print(y.strip('-'))
print(y.lstrip('-'))      # 2). "lstrip" this function is used to strip left side of string..
print(y.rstrip('-'))      # 3). "rstrip" this function is used to strip right side of string..

# "endswith()" and "startswith()" function
```
