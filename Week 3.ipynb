#**Lecture 10:Tutorial on for loop and difference between while loop and for loop.**

```python
# In this lecture we will try to convert this particular while loop into its equivalent for loop.

# Problem 1: Find the Factorial of given number.

""" The snippet we used to solve factorial using while loop is ....

   num = int(input("Enter a number: "))
   fact = 1
   if (num < 0):
    print("Not defined")
   else:
    while (num > 0):
      fact = fact * num
      num = num - 1
    print(fact)                         """

# Let's try to solve this problem using for loop....

num = int(input("Enter a number: "))
fact = 1
if (num < 0):
  print("Not defined")
else:
  for i in range(num, 1, -1):
    fact = fact * i
  print(fact)

""" As you can see, both programs using while loop as well as for loop looks identical,but they have a little bit of difference, as a programmer we have to decide which loop
is more ideal for this particular example.

Now, the question is how to decide that what are the factors which we should consider before deciding whether to use while loop or for loop?
So, the most important factor we should consider is whether do we know the range of the loop are not, as in are we aware in advance that the loop is going to run for some n number
of times, if we are certain that the given loop is going to execute for let us say 10 times or 12 times then we should opt for "for" loop. Whereas, if we do not know how many number
of times the particular loop is going to execute, we should opt for "while" loop. This is important because whenever we use "for" loop we have to provide this range function
and inside that range function, we should also mention the starting point of the loop and the ending point of the loop. In order to provide these counts, we should know how many
number of times the loop will execute.

Hence, for is more suitable when we know the total number of iterations required in that particular program, whereas while is more generic way of execution.

As we can see, this is a problem statement where we are trying to find the factorial of a number and in such a case, the number of iterations are known to us.

Hence, for loop is more suited option than a while loop."""
```

# **Lecture 13: break, continue and pass.**

```python
# Let's take an example to understand these three concepts of break, continue and pass...

# let's think of that we have an email ID which have some userinfo like thier name. we want to print the names only. so, how we gonna do it
# we use a function called break

email = input("Enter your email:\n")
for c in email:
  if (c == '@'):
    break                             # In this snippet break function breaks the loop.
  print(c, end = '')

""" The break function can also be called exit loop function. so, how this piece of snippet function.
first we have a variable called "email" which takes input from the user after that we introduced a foreach function which checks every value
in email for "c". after that we created an if loop in which "c == '@'" that means for loop check this criteria in range 'email' till c equals
to @ after we exit the loop after we find that criteria by using 'break' fuction.  """
```

```python
# Let's see "another function called 'continue'...
email = input("Enter your email: ")
for c in email:
  if(c == "@"):
    print("")
    continue
  print(c, end = "")
```
